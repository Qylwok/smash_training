# Exploring Data

## Initialization

```python
import pandas as pd
from IPython.core.display import HTML

df = pd.read_csv("./csv/pichu_lv9.csv", index_col=False)
df.head()
```

## Sorting by win rate

```python
df["playcount"] = df["win"] + df["lose"]
df["win_rate"] = round(df["win"] / (df["win"] + df["lose"])*1000)/10

display(HTML(df.sort_values(by=["win_rate", "opponent"], ascending=False).dropna().to_html()))
```

## Most encountered

```python
display(HTML(df.sort_values(by=["playcount", "opponent"], ascending=False).to_html()))
# display(df.sort_values(by="playcount", ascending=False))
```

## Global win rate & playcount

### Global win rate

```python
print(df["win"].sum() / df["playcount"].sum() * 100, "%")
```

### Global playcount

```python
print(df["playcount"].sum(), "games")
```
